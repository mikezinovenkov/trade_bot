```python
import numpy as np
import pandas as pd
from catboost import CatBoostClassifier, Pool
from sklearn.metrics import precision_score
model = CatBoostClassifier(loss_function = 'Logloss', eval_metric = 'F1', custom_metric = ['F1', 'Precision', 'Recall'], thread_count=7)
```

```python
symbol = "ALPACA"
time_interval = '1m'
```

```python
crypto_data = pd.read_csv(f'data/training_data_{symbol}_{time_interval}.csv')
```

```python
# создаем новый датафрейм, в которм нет ненужных столбцов
x = crypto_data.drop(['TARGET'], axis=1)
y = crypto_data.TARGET
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=42)
```

```python
model = CatBoostClassifier(depth = 10, l2_leaf_reg = 3, iterations = 250, learning_rate = 0.3)
```

```python
model.fit(x_train, y_train)
```
